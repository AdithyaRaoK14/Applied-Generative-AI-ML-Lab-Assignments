**6. WAP to perform bubble sort, quick sort and insertion sort using functions**

```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Test arrays
arr = [89,14,21,34,55,3]

# Testing all sorting algorithms
print("Bubble Sort Result:", bubble_sort(arr.copy()))
print("Quick Sort Result:", quick_sort(arr.copy()))
print("Insertion Sort Result:", insertion_sort(arr.copy()))
```

**8. Rewrite your pay compulation with time-and-a-half for overtime and create a function called computepay which takes two parameters(hours and rate).**

   **Enter Hours: 45**

   **Enter Rate: 10**

   **Pay: 475.0**              
   
   **475 = 40 * 10 + 5 * 15**

```python
def computepay(hours, rate):
    # Check if hours worked are greater than 40
    if hours > 40:
        # Regular pay for the first 40 hours
        regular_pay = 40 * rate
        # Overtime pay for hours worked beyond 40 (time-and-a-half)
        overtime_pay = (hours - 40) * (rate * 1.5)
        # Total pay
        total_pay = regular_pay + overtime_pay
    else:
        # If no overtime, pay is just hours * rate
        total_pay = hours * rate

    return total_pay

# Prompt the user for input
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

# Calculate and print the total pay
pay = computepay(hours, rate)
print(f"Pay: {pay}")
```

1. Implementation of the classic quicksort algorithm in Python.


```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

arr = [3, 6, 8, 10, 1, 2, 1]
print(quicksort(arr))
```
